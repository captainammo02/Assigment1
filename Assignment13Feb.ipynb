```python
Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.
```

In Python, the Exception class is the base class for all exceptions. When we create a custom exception, we need to inherit from the Exception class so that our custom exception can inherit all the attributes and methods of the base Exception class.

Here are a few reasons why we use the Exception class while creating a custom exception:

1. Inheriting from the Exception class ensures that our custom exception has all the standard exception features like a message string, traceback information, and other attributes.

2. By inheriting from the Exception class, we can easily catch our custom exception along with other built-in exceptions. This makes it easier to handle exceptions in a unified manner in our code.

3. Inheriting from the Exception class also allows our custom exception to be compatible with the raise statement and other exception handling mechanisms in Python.

```python
class MyCustomException(Exception):
    pass

try:
    raise MyCustomException("This is a custom exception")
except MyCustomException as e:
    print(e)
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

The ArithmeticError class is a built-in exception class in Python that is the base class for all arithmetic-related errors. This includes errors such as division by zero, overflow, underflow, and invalid operations on numeric types.

```python
# ZeroDivisionError: This error is raised when a division operation is attempted with a divisor of zero. For example:
a = 10
b = 0
c = a / b  # Raises ZeroDivisionError
```

```python
# OverflowError: This error is raised when the result of an arithmetic operation is too large to be represented by the given data type.
# For example:
import sys
a = sys.maxsize  # Maximum value of a signed integer on this platform
b = a * a        # Raises OverflowError
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is a built-in exception class in Python that is the base class for all lookup-related errors. This includes errors such as when a key or index is not found in a dictionary or sequence, respectively.

```python
# KeyError: This error is raised when a key is not found in a dictionary. For example:
d = {'a': 1, 'b': 2, 'c': 3}
value = d['d']  # Raises KeyError
```

```python
# IndexError: This error is raised when an index is out of range in a sequence. For example:
a = [1, 2, 3]
value = a[3]  # Raises IndexError
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?
```

ImportError is a built-in exception class in Python that is raised when a module, which is needed for an import statement, cannot be found or imported correctly. This can happen for a variety of reasons, such as a typo in the module name, the module not being installed, or the module being located in the wrong directory.

ModuleNotFoundError is a subclass of ImportError that was introduced in Python 3.6. It is specifically raised when a module cannot be found during an import statement. Prior to Python 3.6, an ImportError was raised for both missing modules and other import-related errors.

```python
try:
    import my_module
except ImportError:
    print("Could not import my_module")
```
